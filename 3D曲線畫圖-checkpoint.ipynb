```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
t=np.linspace(-100*np.pi,100*np.pi,100)

x=np.sin(3*t)
y=np.sin(t)+np.cos(t)
z=np.sin(t)+np.cos(t)+np.tan(t)

ax=plt.gca(projection='3d')
plt.plot(x,y,z);
```

```python
x_1=np.sin(t)
y_1=np.cos(87*t)
z_1=np.tan(81*t)

ax=plt.gca(projection='3d')
plt.plot(x_1,y_1,z_1);
```
