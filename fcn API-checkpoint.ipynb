```python
import matplotlib.pyplot as plt
import numpy as np

# Layers for FNN
from tensorflow.keras.models import Sequential,Model
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Flatten

# Layers for CNN
from tensorflow.keras.layers import Dense, Flatten, Conv2D, MaxPool2D, Input, concatenate, add

from tensorflow.keras.optimizers import SGD, Adam

# For data preprocessing
from tensorflow.keras.datasets import fashion_mnist
from tensorflow.keras.utils import to_categorical
```

```python
(X_train, y_train0), (X_test, y_test0) = fashion_mnist.load_data()

X_train = X_train.reshape(-1, 28*28)
X_test = X_test.reshape(-1, 28*28)
X_train = X_train / X_train.max()
X_test = X_test / X_test.max()

y_train = to_categorical(y_train0, 10)
y_test = to_categorical(y_test0, 10)
```

```python
x=Input(shape=(784,))
f_1 = Dense(512, activation='sigmoid')
f_2 = Dense(256, activation='relu')
f_3 = Dense(128,activation='sigmoid')
f_4 = Dense(64,activation='relu')
f_5 = Dense(10,activation='softmax')
```

```python
from IPython.display import Image
Image(filename= "API.jpg")
```

```python
h_1 = f_1(x)
h_2 = f_2(h_1)
h_3 = f_3(h_1)
h_4 = f_4(h_1)
u = concatenate([h_2,h_3,h_4])
v = f_5(u)
```

```python
model=Model(x,v)
model.summary()
```

```python
model.compile(loss='mse', optimizer=Adam(), metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, batch_size=128, epochs=50)
```

```python
score = model.evaluate(X_train, y_train, batch_size=10000)
print(f'Loss: {score[0]}')
print(f'準確率: {score[1]*100}')
```
